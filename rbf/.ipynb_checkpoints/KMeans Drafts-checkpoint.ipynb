```python
#Draft 1
class KMeansClustering:
    def __init__(self):
        self.X = np.random.uniform(-1.0,1.0,(100,2))
        score,uks,sks = self.Iter_KMeansClustering(10)
    def getUk(self,Sk):
        if list(Sk) == []: return np.random.uniform(-1.0,1.0,2)
        return np.average(Sk,axis=0)
    
    def getUks(self,Sks):
        return [self.getUk(Sk) for Sk in Sks]
    
    def removeXnFromSk(self,xn,Sks):
        return [[x for x in Sk if (x != xn).all()] for Sk in Sks]
    
    def indxOfXnInSks(self,xn,Sks):
        for indx, Sk in enumerate(Sks):
            if len([ True for sk in Sk if (sk == xn).all()]) != 0:
                return indx
            
    def changeSks(self,Sks,Uks):
        numChange = 0
        for xn in self.X:
            oldSkIndx = self.indxOfXnInSks(xn,Sks)
            distances = [np.linalg.norm(uk-xn)for uk in Uks]
            bestSkIdx = distances.index(min(distances))
            if oldSkIndx != bestSkIdx:
                Sks = self.removeXnFromSk(xn,Sks)
                Sks[bestSkIdx] = Sks[bestSkIdx]+[xn]
                numChange += 1
            
        return Sks,numChange
    
    def initializeSks(self,numUks):
        X = copy.deepcopy(self.X)
        np.random.shuffle(X)
        Sks = np.array_split(self.X,numUks) 
        return Sks
    
    def removeEmpty(self, Sks):
        return [Sk for Sk in Sks if Sk != []]
    def empty(self, Sks):
        return len([Sk for Sk in Sks if Sk == []])>0
    
    def Iter_KMeansClustering(self,numUks):
        Sks = self.initializeSks(numUks)
        numChange = 100
        while numChange > 0:
            Uks = self.getUks(Sks)#Sk is fixed
            Sks,numChange = self.changeSks(Sks,Uks) #move xn to best Sk
            print(sum([len(x) for x in Sks]))
            #if self.empty(Sks):
                #numChange=100
                #Sks = self.initializeSks(numUks)
        
        Uks = self.getUks(Sks)
        score = self.MinimizeLloyedAlgorithm(Sks,Uks)
        return score, Uks, Sks
    
    def runKMeansClustering(self,numUks):
        lowestScore = 1000
        for x in range(1):
            score, Uks, Sks = self.Iter_KMeansClustering(numUks)
            if lowestScore > score:
                lowestScore = score
                self.Uks = Uks
                self.Sks = Sks
        print(lowestScore)
        print(self.Uks)
        
    
        
    def MinimizeLloyedAlgorithm(self,Sks,Uks):
        return np.sum(np.sum([[np.linalg.norm(xn-uk,ord=2) for xn in Sk] for uk, Sk in zip(Uks,Sks)]))
            
        
            
```

```python
#Draft 2
#Loading the required modules
 
import numpy as np

class kmeans():
    def __init__(self,X):
        self.X = X
        self.k = 10
        self.runKMeansClustering()
        
    def runKMeansClustering(self):
        lowestScore = 100
        for x in range(10):
            score, Uks = self.Iter_KMeansClustering()
            if lowestScore > score:
                lowestScore = score
                self.Uks = Uks
        print(lowestScore)
        print(self.Uks)
        
    def Iter_KMeansClustering(self):
        Uks = np.random.uniform(-1.0,1.0,(10,2))
        oldUks = Uks
        distances = self.getDistance(Uks)
        points = np.array([np.argmin(i) for i in distances])
        while len(np.unique(points))!=self.k:
                Uks = np.random.uniform(-1.0,1.0,(10,2))
                points = self.AdjustSks(Uks)
        while True:
            
            Uks = self.getUks(points)
                
            if np.equal(oldUks,Uks).all():
                break

            

            distances = self.getDistance(Uks)
            points = np.array([np.argmin(i) for i in distances])
            
            while len(np.unique(points))!=self.k:
                Uks = np.random.uniform(-1.0,1.0,(10,2))
                points = self.AdjustSks(Uks)
                
            oldUks = copy.deepcopy(Uks)
            

        score = self.MinimizeLloyedAlgorithm(points,Uks)
        return score,Uks
    
    def AdjustSks(self,Uks):
        distances = self.getDistance(Uks)
        points = np.array([np.argmin(i) for i in distances])
        return points
    
    def getUks(self,points):
        Uks = []
        for k in range(self.k):
            Sk = self.getSk(points,k)
            uk = Sk.mean(axis=0) 
            Uks.append(uk)
        return Uks
    
    def getSk(self,points,k):
        return self.X[points==k]
    
    def getDistance(self,Uks):
        return [[np.linalg.norm(xn-uk) for uk in Uks] for xn in self.X]
    
    def MinimizeLloyedAlgorithm(self,points,Uks):
        Sks = [self.getSk(points,k) for k in range(self.k)]
        return np.sum(np.sum([[np.linalg.norm(xn-uk,ord=2) for xn in Sk] for uk, Sk in zip(Uks,Sks)]))
 
```

```python
#Final
#Loading the required modules
 
import numpy as np

class kmeans():
    def __init__(self,X):
        self.X = X
        self.k = 10
        self.runKMeansClustering()
        
    def runKMeansClustering(self):
        lowestScore = 100
        for x in range(10):
            score, Uks = self.Iter_KMeansClustering()
            if lowestScore > score:
                lowestScore = score
                self.Uks = Uks
        print(lowestScore)
        print(self.Uks)
    
    def empty(self, Sks):
        return len([Sk for Sk in Sks if Sk == []])>0
    
    def Iter_KMeansClustering(self):
        Uks = np.random.uniform(-1.0,1.0,(10,2))
        oldUks = Uks
        Sks = self.AdjustSks(Uks)
        
        while self.empty(Sks):
                Uks = np.random.uniform(-1.0,1.0,(10,2))
                Sks = self.AdjustSks(Uks)
                
        while True:
            
            Uks = self.getUks(Sks)
                
            if np.equal(oldUks,Uks).all():
                break

            

            Sks = self.AdjustSks(Uks)
            
            while self.empty(Sks):
                Uks = np.random.uniform(-1.0,1.0,(10,2))
                Sks = self.AdjustSks(Uks)
                
            oldUks = copy.deepcopy(Uks)
            

        score = self.MinimizeLloyedAlgorithm(Sks,Uks)
        return score,Uks
    
    def AdjustSks(self,Uks):
        Sks = [[] for x in range(self.k)]
        distances = self.getDistance(Uks)
        for xn,i in zip(self.X,distances):
            Sks[np.argmin(i)] = Sks[np.argmin(i)]+[xn]
        return Sks
    
    def getUks(self,Sks):
        Uks = []
        for k in range(self.k):
            Sk = self.getSk(Sks,k)
            uk = np.array(Sk).mean(axis=0) 
            Uks.append(uk)
        return Uks
    
    def getSk(self,Sks,k):
        return Sks[k]
    
    def getDistance(self,Uks):
        return [[np.linalg.norm(xn-uk) for uk in Uks] for xn in self.X]
    
    def MinimizeLloyedAlgorithm(self,Sks,Uks):
        return np.sum(np.sum([[np.linalg.norm(xn-uk,ord=2) for xn in Sk] for uk, Sk in zip(Uks,Sks)]))
 
```
